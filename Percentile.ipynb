```python
# retrun statement is must for function and alignment also properly align
def QuanQual(dataset):
    Qual = []
    Quan = []
    for ColumnName in dataset.columns:
       # print(ColumnName)
        if(dataset[ColumnName].dtype =='O'):
            #print("Qual")
            Qual.append(ColumnName)
            #print(Qual)
        else:
           # print("Quan")
            Quan.append(ColumnName)
            #print(Quan)
    return Qual, Quan
```

```python
QuanQual(dataset)
```

```python
Qual, Quan = QuanQual(dataset)
```

# Percentile

```python
import numpy as np
```

```python
np.percentile(dataset["ssc_p"],20)
```

```python
Descriptive = pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%"],columns=Quan)
for ColumnName in Quan:
    Descriptive[ColumnName]["Mean"] = dataset[ColumnName].mean()
    Descriptive[ColumnName]["Median"] = dataset[ColumnName].median()
    Descriptive[ColumnName]["Mode"] = dataset[ColumnName].mode() [0]
   # The below scripts shows NaN value in Salary column because Salary column is contains null data
    Descriptive[ColumnName]["Q1:25%"] = np.percentile(dataset[ColumnName],25)
    Descriptive[ColumnName]["Q2:50%"] = np.percentile(dataset[ColumnName],50)
    Descriptive[ColumnName]["Q3:75%"] = np.percentile(dataset[ColumnName],75)
    Descriptive[ColumnName]["99%"] = np.percentile(dataset[ColumnName],99)
    Descriptive[ColumnName]["Q4:100%"] = np.percentile(dataset[ColumnName],100)
```

```python
Descriptive
```

```python
Descriptive = pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%"],columns=Quan)
for ColumnName in Quan:
    Descriptive[ColumnName]["Mean"] = dataset[ColumnName].mean()
    Descriptive[ColumnName]["Median"] = dataset[ColumnName].median()
    Descriptive[ColumnName]["Mode"] = dataset[ColumnName].mode() [0]
    Descriptive[ColumnName]["Q1:25%"] = dataset.describe()[ColumnName],["25%"]
    Descriptive[ColumnName]["Q2:50%"] = dataset.describe()[ColumnName],["50%"]
    Descriptive[ColumnName]["Q3:75%"] = dataset.describe()[ColumnName],["75%"]
    Descriptive[ColumnName]["99%"] = np.percentile(dataset[ColumnName],99)
    Descriptive[ColumnName]["Q4:100%"] = dataset.describe()[ColumnName],["max%"] 
    
```
